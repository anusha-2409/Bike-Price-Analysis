```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
HEADERS = {"User-Agent": "Mozilla/5.0"}
TOTAL_PAGES = 50
PAGE_DELAY = 2
DETAIL_DELAY = 2
bike_name = []
model_year = []
price = []
location = []
kms_run = []
color = []
engine_cc = []
no_of_owners = []

for page in range(1, TOTAL_PAGES + 1):
    url = f"https://www.bikes4sale.in/used/buy/?page={page}"
    response = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(response.text, "html.parser")

    bikes = soup.find_all('div', class_='txtBox')
    if not bikes:
        print(f"No bikes found on page {page}")
        break

    for x in bikes:
        text_lines = list(x.stripped_strings)
        full_name = text_lines[0]
        parts = full_name.split(',')
        bname = parts[0].strip()
        m_year = parts[1].strip().split()[0] if len(parts) > 1 else ""
        loc = text_lines[1].replace('Location:', '').strip() if len(text_lines) > 1 else ""
        price_tag = x.find('span', class_='prz')
        pr = price_tag.text.strip() if price_tag else ""

        km = ""
        clr = ""
        eng_cc = ""
        owners = ""
        reason = ""

        detail_tag = x.find("a", string="View More Details")
        if detail_tag:
            detail_url = detail_tag["href"]
            if not detail_url.startswith("http"):
                detail_url = "https://www.bikes4sale.in" + detail_url

            try:
                dpage = requests.get(detail_url, headers=HEADERS)
                if dpage.status_code == 200:
                    dsoup = BeautifulSoup(dpage.text, "html.parser")
                    rows = dsoup.select("table.table.idNewTbl tr")
                    for row in rows:
                        cols = row.find_all("td")
                        if len(cols) >= 2:
                            label = cols[0].get_text(strip=True).lower()
                            value = cols[1].get_text(strip=True)
                            if "kms run" in label:
                                km = value
                            elif "colour" in label:
                                clr = value
                            elif "engine cc" in label:
                                eng_cc = value
                            elif "no of owners" in label:
                                owners = value

                time.sleep(random.uniform(DETAIL_DELAY, DETAIL_DELAY + 1))
            except:
                pass
        bike_name.append(bname)
        model_year.append(m_year)
        price.append(pr)
        location.append(loc)
        kms_run.append(km)
        color.append(clr)
        engine_cc.append(eng_cc)
        no_of_owners.append(owners)

    print(f"Page {page} scraping completed!")
    time.sleep(random.uniform(PAGE_DELAY, PAGE_DELAY + 1))

df = pd.DataFrame({
    "Bike Name": bike_name,
    "Model Year": model_year,
    "Price": price,
    "Location": location,
    "KMs Run": kms_run,
    "Color": color,
    "Engine CC": engine_cc,
    "No of Owners": no_of_owners,

})

df.to_csv("bikes4sale_Dataset_Final.csv", index=False)
print(f"Scraping completed! Total records: {len(df)}")
```
